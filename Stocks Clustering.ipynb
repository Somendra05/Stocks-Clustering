```python
import yfinance as yf
import pandas as pd

# List of stock tickers (you can modify this list)
tickers = ['AAPL', 'MSFT', 'GOOGL', 'AMZN', 'TSLA', 'META', 'NVDA']

# Time period for historical data
start_date = '2023-01-01'
end_date = '2024-01-01'

# Download data
stock_data = yf.download(tickers, start=start_date, end=end_date)

# Save to CSV
stock_data.to_csv('/Users/som/Downloads/stock_data.csv')

# Preview data
print(stock_data.head())
```

## Data Preprocessing

```python
stock_data.info()
```

```python
stock_data.head()
```

```python
# Handle missing values (if any)
stock_data = stock_data.dropna()

# Verify the cleaned data
print(stock_data.isnull().sum())
```

## Feature Engineering

```python
# Use the 'Close' price for calculations
close_prices = stock_data['Close']

# Calculate daily returns
returns = close_prices.pct_change()

# Calculate moving averages
moving_avg_50 = close_prices.rolling(window=50).mean()
moving_avg_200 = close_prices.rolling(window=200).mean()

# Calculate volatility (Standard Deviation of returns)
volatility = returns.rolling(window=30).std()

# Calculate cumulative returns
cumulative_returns = (1 + returns).cumprod()

# Combine features into a new dataframe
features = pd.DataFrame({
    'Return': returns.mean(),
    'Volatility': volatility.mean(),
    'Moving_Avg_50': moving_avg_50.mean(),
    'Moving_Avg_200': moving_avg_200.mean(),
    'Cumulative_Returns': cumulative_returns.iloc[-1]
})

# Drop any rows with missing values
features = features.dropna()

# Preview the feature data
print(features)
```

## Normalize Data for Clustering

```python
from sklearn.preprocessing import StandardScaler

# Scale the features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Convert scaled data back to DataFrame
scaled_features_df = pd.DataFrame(scaled_features, index=features.index, columns=features.columns)
print(scaled_features_df.head())
```

## K-means Clustering

```python
from sklearn.cluster import KMeans
import matplotlib
import matplotlib.pyplot as plt

# Try different k values (up to the number of stocks)
wcss = []

for k in range(1, len(features) + 1):  # Ensure k doesn't exceed number of samples
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(scaled_features_df)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Curve
plt.plot(range(1, len(features) + 1), wcss)
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method for Optimal K')
plt.show()
```

```python
# Apply KMeans with the optimal k
optimal_k = 3

kmeans = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)
clusters = kmeans.fit_predict(scaled_features_df)

# Add cluster labels to the original features dataframe
features['Cluster'] = clusters
print(features)
```

## PCA 
We’ll use PCA to reduce dimensions and visualize the clusters.

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_features_df)

plt.figure(figsize=(10, 6))
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=clusters, cmap='viridis')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.title('Stock Clustering Visualization (PCA)')
plt.colorbar(label='Cluster')
plt.show()
```

```python
# Group by cluster and calculate mean
cluster_summary = features.groupby('Cluster').mean()
print(cluster_summary)
```

## Hierarchical Clustering

```python
from scipy.cluster.hierarchy import dendrogram, linkage

# Perform hierarchical clustering
linkage_data = linkage(scaled_features_df, method='ward')
dendrogram(linkage_data, labels=features.index)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Stocks')
plt.ylabel('Distance')
plt.show()
```

Dendrogram Analysis:
1. Cluster Interpretation:

    The dendrogram shows three main clusters:
- Cluster 1: MSFT (Microsoft), AAPL (Apple), AMZN (Amazon), GOOGL (Google) – Companies with similar market behavior and stability.
- Cluster 2: NVDA (Nvidia) and META (Meta) – More volatile and high-growth tech companies.
- Cluster 3: TSLA (Tesla) – Stands apart, likely due to its high volatility and unique growth trajectory.

2. Distance Interpretation:
- The vertical distance at which clusters merge indicates the similarity between the stocks.
- MSFT and AAPL merge at a short distance, indicating high similarity.
- TSLA merges later, suggesting it behaves very differently from the rest of the stocks.

3. Insights:
- This confirms that TSLA and NVDA are likely high-risk, high-reward stocks.
- MSFT, AAPL, AMZN, and GOOGL belong to a more stable cluster, indicating lower volatility and steady returns.
- META sits between stability and high volatility.

- Cumulative Returns vs. Moving_Avg_50/200 (-0.32):This could indicate that stocks showing strong cumulative returns may not conform to traditional long-term support levels.

3. Sharpe Ratio and Performance:
- Sharpe Ratio and Cumulative Returns (0.78): A high Sharpe ratio correlates well with cumulative returns, reinforcing that stocks with favorable risk-adjusted returns often exhibit strong long-term performance.
